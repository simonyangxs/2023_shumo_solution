```python
import pandas as pd
from tqdm import *
import os
import math
from datetime import timedelta
```

```python
tab1 = pd.read_excel('metadata/表1-患者列表及临床信息.xlsx')
tab1 = tab1.rename(columns={'Unnamed: 0': 'ID'})
```

```python
tab1['发病到首次影像检查时间间隔'].describe()
```

```python
tab2_dict = {}
for i, sheet_name in enumerate(['首次', '随访1', '随访2', '随访3', '随访4', '随访5', '随访6', '随访7', '随访8']):
    if sheet_name == '首次':
        col = '首次检查流水号'
    else:
        col = sheet_name + '流水号'

    tab2_dict[i] = pd.read_excel('metadata/表2-患者影像信息血肿及水肿的体积及位置.xlsx', sheet_name=sheet_name, usecols=['ID', col, 'HM_volume'])
```

是否发生血肿扩张可根据血肿体积前后变化，具体定义为：后续检查 比首次检查绝对体积增加≥6 mL 或相对体积增加≥33%

```python
'''计算在哪一次开始发生血肿扩张'''
tab2_cmp = pd.DataFrame()
tab2_cmp['ID'] = tab2_dict[0]['ID']
tab2_cmp['num_0'] = tab2_dict[0]['首次检查流水号']  # 流水号
tab2_cmp['vol_0'] = tab2_dict[0]['HM_volume'] / 1000 # 血肿总体积（ml）

for k in tab2_dict.keys():
    if k == 0:
        continue

    tab2_cmp['num_%d' % k] = tab2_dict[k]['随访%d流水号' % k] # 流水号
    tab2_cmp['vol_%d' % k] = tab2_dict[k]['HM_volume'] / 1000 # 血肿总体积（ml）
    tab2_cmp['abs_%d' % k] = tab2_cmp['vol_%d' % k] - tab2_cmp['vol_0']  # 血肿绝对体积增加
    tab2_cmp['rela_%d' % k] = tab2_cmp['vol_%d' % k] / tab2_cmp['vol_0'] - 1 # 血肿相对体积增加

    tab2_cmp['if_HM_%d' % k] = tab2_cmp.apply(lambda x: 1 if x['abs_%d' % k] >= 6 or x['rela_%d' % k] >= 0.33 else 0, axis=1) # 判断绝对体积增加是否超过6ml或相对体积增加超过33%
```

```python
tab2_cmp = pd.merge(tab2_cmp, tab1[['ID', '发病到首次影像检查时间间隔']], on='ID', how='left') # 补充发病到首次影像检查时间间隔的数据
```

```python
'''补充流水号与时间的对应'''
ann_tab = pd.read_excel('metadata/附表1-检索表格-流水号vs时间(1).xlsx')
```

```python
ann_dict = {} # 获得一个流水号-时间对应的字典
col_list = ann_tab.columns.tolist()
for col in col_list:
    if '流水号' in col:
        print(col, col_list[col_list.index(col) - 1])
        for num, time in zip(ann_tab[col], ann_tab[col_list[col_list.index(col) - 1]]):
            if math.isnan(num): # 无随访记录
                continue
            ann_dict[num] = time
```

```python
'''检查一下流水号字典和所有患者的流水号是否对应'''
num_in_tab2 = []
num_in_ann = list(ann_dict.keys())

for col in tab2_cmp.columns:
    if 'num' in col:
        num_in_tab2 += [x for x in tab2_cmp[col] if not math.isnan(x)]
```

```python
set(num_in_tab2) - set(num_in_ann) # 表2记录中的所有流水号均在附件2中出现
```

```python
set(num_in_ann) - set(num_in_tab2) # 附件2中记录的一些流水号没有在表2中出现，合理（附件2中最多记录到了第13次随访，而表2最多只有8次）
```

```python
'''获取每个人每次体检到首次影像时间'''
for col in tab2_cmp.columns:
    if 'num' in col:
        i = int(col.replace('num_', ''))
        tab2_cmp['time_%d' % i] = tab2_cmp[col].map(lambda x: ann_dict[x] if x in ann_dict.keys() else None)

        if i != 0:  
            # 计算当次随访时间到发病时间的间隔（注意：是当次随访时间-首次检查时间+发病到首次影像检查时间间隔）
            # 转换时间格式
            tab2_cmp['interval_%d' % i] = tab2_cmp.apply(lambda x: (x['time_%d' % i] - x['time_0']).total_seconds() / 3600 + x['发病到首次影像检查时间间隔'], axis=1)
```

```python
tab2_cmp.to_excel('processed/Q1a_record.xlsx', index=False)
```

```python
'''
保留最主要的列
- 当次检测结果是否血肿扩张
- 当次检测时间到发病时间的间隔
'''
tab2_HM = tab2_cmp[['ID'] + ['if_HM_%d' % i for i in range(1, 9)] + ['interval_%d' % i for i in range(1, 9)]][:100] # 只需计算前100人
```

```python
tab2_HM['是否发生血肿扩张'] = None
tab2_HM['血肿扩张时间'] = None

for i, row in tab2_HM.iterrows():
    for j in range(1, 9):
        if row['if_HM_%d' % j] == 1 and row['interval_%d' % j] <= 48:
            tab2_HM.loc[i, '是否发生血肿扩张'] = 1
            tab2_HM.loc[i, '血肿扩张时间'] = row['interval_%d' % j]
            break

        else:
            tab2_HM.loc[i, '是否发生血肿扩张'] = 0
```
