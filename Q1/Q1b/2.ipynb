```python
from pycaret.datasets import get_data

import pickle
def load_pkl(path):
    with open(path, 'rb') as f:
        return pickle.load(f)
from pycaret.classification import *
import pandas as pd
random_state=42
name= '2'
```

```python
juice = pd.read_csv(f'./data{name}.csv')
exp = ClassificationExperiment()
```

```python
train_df = juice.head(100)
test_df = juice.tail(60)
```

```python
exp_name = exp.setup(data = train_df,  target = 'label')
exp.set_config('seed', random_state)
```

```python
exp.models()
```

```python
top5 = exp.compare_models(sort = 'acc',n_select = 5,budget_time=360)
```

```python
for best_model in top5:
    result = exp.predict_model(best_model,data=test_df.iloc[:,:-1])# generate predictions
    result = result.reset_index()
    result['label'] = list(test_df['label'])
    result['target'] = (result['prediction_label'] == result['label'])
    print(result['target'].mean())
```

```python
exp.save_model(top5, f'./result/{name}')
```
