```python
from pycaret.datasets import get_data

import pickle
def load_pkl(path):
    with open(path, 'rb') as f:
        return pickle.load(f)
from pycaret.classification import *
import tqdm
import pandas as pd
name= '5'

juice = pd.read_csv(f'./data{name}.csv')
exp = ClassificationExperiment()

train_df = juice.head(100)
test_df = juice.tail(60)
```

```python
top5list = {}
for random_state in tqdm.tqdm(range(0,101,4)):
    exp.set_config('seed', random_state)
    exp_name = exp.setup(data = train_df,  target = 'label',verbose=False)
    top5 = exp.compare_models(sort = 'auc',n_select = 5,budget_time=360,verbose=True)
    top5list[random_state] = [top5,exp._display_container[1] ] 
```

```python
random_state=42
exp.set_config('seed', random_state)
exp_name = exp.setup(data = train_df,  target = 'label',verbose=False)
top5 = exp.compare_models(sort = 'auc',n_select = 5,budget_time=360,verbose=True)
top5list[random_state] = [top5,exp._display_container[1] ] 
```

```python
exp.predict_model(top5list[20][0][0],data=test_df.iloc[:,:-1]).prediction_score.value_counts()# generate predictions
```

```python
def save_pkl(path,obj):
    with open(path, 'wb') as f:
        pickle.dump(obj,f)
        
def load_pkl(path): 
    with open(path, 'rb') as f:
        return pickle.load(f) 
save_pkl(f'{name}.pkl',top5list)
```

```python
all_result = pd.DataFrame()
for i in top5list.keys():
    t = top5list[i][1]
    t['random'] = i
    all_result=all_result.append(t)
```
